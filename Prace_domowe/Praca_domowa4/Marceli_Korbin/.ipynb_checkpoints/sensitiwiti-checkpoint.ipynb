## Oryginalna sieć neuronowa

Buduję sieci neuronowe o 5 warstwach ukrytych po 8 neuronów każda. Tak jak w poprzedniej pracy domowej, wykorzystywanymi funkcjami aktywacji będą sigmoidalna w warstwach ukrytych i softmax na wyjściowej. Zachowujemy również *categorical_crossentropy*, Adama i metrykę accuracy. Za każdym razem sieci będą uczone przez 400 epok z 32 elementami w każdym batchu.

Tak jak poprzednio, wykorzystuję zbiór *rings5-regular*.

```python
import keras, numpy as np, pandas as pd, seaborn as sns, matplotlib.pyplot as plt, copy
from keras.models import Sequential
from keras.layers import Dense
from keras import activations
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score

rings_train = pd.read_csv('rings5-regular-training.csv')
rings_test = pd.read_csv('rings5-regular-test.csv')
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='x', y='y', hue='c', data=rings_train, legend="full")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='x', y='y', hue='c', data=rings_test, legend="full")
```

```python
X_train = rings_train.iloc[:,:2].values
Y_train = rings_train.iloc[:,2:3].values
X_test = rings_test.iloc[:,:2].values
Y_test = rings_test.iloc[:,2:3].values
koder = OneHotEncoder()
Y_train = koder.fit_transform(Y_train).toarray()
Y_test = koder.fit_transform(Y_test).toarray()
```

```python
def normal_learn():
    M = Sequential()
    M.add(Dense(8, input_dim=2, activation='sigmoid'))
    for x in range(4):
        M.add(Dense(8, activation='sigmoid'))
    M.add(Dense(5, activation='softmax'))
    M.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    F = M.fit(X_train, Y_train, epochs=400, batch_size=32, verbose=0)
    Y_test0 = M.predict(X_test) # dane przesunięte o parametr shoot na obu osiach!
    predicted = [np.argmax(Y_test0[x]) for x in range(len(Y_test0))]
    actual = [np.argmax(Y_test[x]) for x in range(len(Y_test))]
    return accuracy_score(predicted, actual)*100.0
```

Wyniki uczenia będę przedstawiać w postaci średniej i odchylenia standardowego z wartości miary accuracy dla 10 treningów sieci.

```python
def vect_normal():
    vect = []
    for x in range(10):
        vect.append(normal_learn())
    print('średnia: ', np.mean(vect), 'odchylenie: ', np.std(vect))
```

```python
vect_normal()
```

## Perturbacje w danych wejściowych

Obie współrzędne danych testowych w całym zbiorze przesuwam o jednakową wartość: 0,5, 1, 5 i 10.

```python
def data_pert(shoot):
    M = Sequential()
    M.add(Dense(8, input_dim=2, activation='sigmoid'))
    for x in range(4):
        M.add(Dense(8, activation='sigmoid'))
    M.add(Dense(5, activation='softmax'))
    M.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    F = M.fit(X_train, Y_train, epochs=400, batch_size=32, verbose=0)
    Y_test0 = M.predict(X_test+shoot) # dane przesunięte o parametr shoot na obu osiach!
    predicted = [np.argmax(Y_test0[x]) for x in range(len(Y_test0))]
    actual = [np.argmax(Y_test[x]) for x in range(len(Y_test))]
    return accuracy_score(predicted, actual)*100.0
```

```python
def vect_data(shoot):
    vect = []
    for x in range(10):
        vect.append(data_pert(shoot))
    print('średnia: ', np.mean(vect), 'odchylenie: ', np.std(vect))
```

```python
vect_data(0.5)
```

```python
vect_data(1)
```

```python
vect_data(5)
```

```python
vect_data(10)
```

Już przesunięcie o 0.5 obniżyło skuteczność sieci, aczkolwiek przy większych przesunięciach do pewnego momentu średnia miara _accuracy_ nie spada, a wręcz się stabilizuje. Dopiero przesunięcie o 10 osłabia jakość uczenia.

## Perturbacje w wagach

Na tym etapie przekształcam wszytkie wagi w wybranej warstwie (lub we wszystkich warstwach) modelu. Będę stosować przesuwanie (o 1, 5, 10) lub mnożenie (przez -1, 2, 6).

### Warstwy modyfikowane pojedynczo

```python
def weights_changed(l, func):
    M = Sequential()
    M.add(Dense(8, input_dim=2, activation='sigmoid'))
    for x in range(4):
        M.add(Dense(8, activation='sigmoid'))
    M.add(Dense(5, activation='softmax'))
    M.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
    for layer in l:
        weyts = M.layers[layer].get_weights()
        M.layers[layer].set_weights([func(weyts[0]), func(weyts[1])])
    # namieszanie w wagach
        
    F = M.fit(X_train, Y_train, epochs=400, batch_size=32, verbose=0)
    Y_test0 = M.predict(X_test)
    predicted = [np.argmax(Y_test0[x]) for x in range(len(Y_test0))]
    actual = [np.argmax(Y_test[x]) for x in range(len(Y_test))]
    return accuracy_score(predicted, actual)*100.0
```

```python
def vect_layer(l, func):
    vect = []
    for x in range(10):
        vect.append(weights_changed(l, func))
    print('średnia: ', np.mean(vect), 'odchylenie: ', np.std(vect))
```

#### Różne przekształcenia danych na jednej warstwie

```python
vect_layer([0], lambda A: A*6)
```

Najmniej szkód, a wręcz korzyści, przyniosło pomnożenie wag przez -1. Reszta przekształceń jest coraz bardziej szkodliwa, im bardziej te przekształcenia są gwałtowne. Warto zwrócić uwagę na wysokie odchylenie miary accuracy po pomnożeniu wag przez 6.

Nawet mnożenie wszystkich wartości przez -1 mimo niezłej średniej daje wyniki, które można zinterpretować jako gwarantowany minimalny poziom accuracy sięgający dopiero okolic 70%.

## Zjawisko niepewności

Niepewność w sieci neuronowej dzieli się na dwa rodzaje: **aleatoryczną** i **epistemiczną**.
- Pierwszy rodzaj niepewności wiąże się z danymi wejściowymi i występującym w nich szumie; w związku z nią stosuje się m.in. augmentację danych czy trenowanie modeli w jednej grupie.
- Drugi typ niepewności polega na nieznajomości prawdziwych warunków, które wpłynęły na dane i przejawia się w wartościach parametrów sieci. Szeroko stosowaną metodą powiązaną z niepewnością epistemiczną jest bayesowska sieć neuronowa, która (zgodnie z nazwą) wykorzystuje estymację bayesowską.

źródło: Abbasnejad E., Uncertainity in Machine Learning, _The University of Adelaide_ https://cs.adelaide.edu.au/~javen/talk/ML05_Uncertainty_in_DL.pdf
