Celem pracy jest zgłębienie tematu _ablation study_.

## A co to jest?

* _Ablation study_ polega na usuwaniu elementów modelu uczenia maszynowego, w celu sprawdzenia ich wpływu na jakość procesu uczenia.
* Pojęcie to wywodzi się z biologii i jest inspirowane zabiegiem _ablacji_, czyli usuwania narządu z ciała człowieka, przeprowadzanym dla sprawdzenia sprawności organizmu w jego nieobecności. (Nie mylić z ablacją jako metodą leczenia arytmii.)
* W praktyce, _ablation study_ sprowadza się do obserwacji jakości uczenia w wypadku np. usunięcia kolumny ze zbioru danych, komponentu z modelu lub warstwy z sieci neuronowej.
    * W zależności od typu wycofywanego elementu, można rozpatrywać różne typy tzw. *ablation trials*, czyli prób uczenia modelu bez danego elementu z dostępnych: *model ablation trial*, _feature ablation trial_ czy _layer ablation trial_.

## Bawimy się w ablation study

Zbudujemy 26 modeli: jeden podstawowy (odnośnikowy) i aż 25 pobocznych, na których będziemy sprawdzać wpływ modyfikacji sieci. Będziemy uczyć wszystkie sieci zadania klasyfikacji punktów na płaszczyźnie dwuwymiarowej (obie współrzędne z zakresu od -100 do 100) do jednego z pięciu zbiorów, rozdzielonych w manierze pierścieni.

Zbiór (oryginalnie zatytułowany rings5-regular) zapożyczyłem z zajęć metod inteligencji obliczeniowej w analizie danych.

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
from keras import activations
```

```python
import numpy as np, pandas as pd
rings_train = pd.read_csv('rings5-training.csv')
rings_test = pd.read_csv('rings5-test.csv')
```

```python
import seaborn as sns, matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16, 9))
sns.scatterplot(x='x', y='y', hue='c', data=rings_train, legend="full")
```

```python
X_train = rings_train.iloc[:,:2].values
Y_train = rings_train.iloc[:,2:3].values
X_test = rings_test.iloc[:,:2].values
Y_test = rings_test.iloc[:,2:3].values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score
```

W tym zbiorze trzeba przekodować kolumnę odpowiedzi w trybie one-hot, bo na teraz przyjmuje ona dyskretne wartości od 0 do 4.

```python
koder = OneHotEncoder()
Y_train = koder.fit_transform(Y_train).toarray()
Y_test = koder.fit_transform(Y_test).toarray()
```

### Model podstawowy

Model podstawowy będzie miał 7 warstw o liczbie neuronów kolejno: 15, 14, 13, 12, 11, 10, 5. Ostatnia warstwa będzie używać funkcji aktywacji softmax, a cała reszta - sigmoidalnej. Naszą funkcją straty będzie *categorical\_crossentropy*, optymizatorem Adam, a metryką _accuracy_.

Docelowo, sieć uczymy przez 200 epok z podziałem na 16 podzbiorów (batchów) w każdej z nich. Żeby nie zużywać bez sensu miejsca, przebieg uczenia sieci zostanie zapisany tylko dla podstawowego modelu. Po nauce liczymy metrykę _accuracy_ dla części treningowej i dla części testowej.

```python
M0 = Sequential()
M0.add(Dense(15, input_dim=2, activation='sigmoid'))
for i in range(14, 9, -1):
    M0.add(Dense(i, activation='sigmoid'))
M0.add(Dense(5, activation='softmax'))
M0.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
def uczenie(M, ver=0, ep=200, bs=16):
    F = M.fit(X_train, Y_train, epochs=ep, batch_size=bs, verbose=ver)
    Y_train0 = M.predict(X_train)
    predicted0 = [np.argmax(Y_train0[x]) for x in range(len(Y_train0))]
    actual0 = [np.argmax(Y_train[x]) for x in range(len(Y_train))]
    Y_test0 = M.predict(X_test)
    predicted1 = [np.argmax(Y_test0[x]) for x in range(len(Y_test0))]
    actual1 = [np.argmax(Y_test[x]) for x in range(len(Y_test))]
    print('WYNIK: ', accuracy_score(predicted0, actual0)*100.0, '% na treningowej, ', accuracy_score(predicted1, actual1)*100.0, '% na testowej')
```

```python
uczenie(M0, ver=1)
```

Notujemy: **95,28**% dla części treningowej, **93,6**% dla testowej.

### Usuwanie warstw

Pierwszą modyfikacją będzie usuwanie jednej z warstw ukrytych (o liczbie neuronów między 10 a 14; sieć będzie mieć 6 warstw).

Tworzymy pięć sieci, o podobnej architekturze co w podstawowej sieci, ale po kolei **bez warstwy** o 10, 11, 12, 13, 14 neuronach i notujemy wyniki.

```python
def packmodel(M, nope): # nope - nie chcemy warstwy o ilu neuronach?
    M.add(Dense(15, input_dim=2, activation='sigmoid'))
    for i in range(14, 9, -1):
        if not i == nope:
            M.add(Dense(i, activation='sigmoid'))
    M.add(Dense(5, activation='softmax'))
    M.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

M_out_layer = [Sequential() for x in range(5)]
for x in range(5):
    packmodel(M_out_layer[x], x+10)
    uczenie(M_out_layer[x])
```

Najlepsze wyniki uzyskały sieci bez warstwy o 11 neuronach i bez warstwy o 14 neuronach; ta druga wypadła lepiej na części treningowej zbioru, ale na testowej nieznacznie lepiej wypadła pierwsza z wymienionych.

### Dodawanie warstw

A teraz na odwrót: jedną warstwę o danej liczbie neuronów będziemy dodawać do sieci 2 razy, zatem sieć będzie mieć 8 warstw.

Po kolei duplikujemy warstwę o 10, 11, 12, 13, 14 neuronach.

```python
def packmodel2(M, yep): # yep - którą warstwę podwoić?
    M.add(Dense(15, input_dim=2, activation='sigmoid'))
    for i in range(14, 9, -1):
        M.add(Dense(i, activation='sigmoid'))
        if i == yep:
            M.add(Dense(i, activation='sigmoid'))
    M.add(Dense(5, activation='softmax'))
    M.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

M_in_layer = [Sequential() for x in range(5)]
for x in range(5):
    packmodel2(M_in_layer[x], x+10)
    uczenie(M_in_layer[x])
```

Najlepiej, a ostatecznie wręcz lepiej niż oryginalna, spisała się sieć o architekturze z liczbami neuronów: 15-14-13-12-11-11-10-5.

### Zmiana funkcji aktywacji (jako parametru warstwy) na ReLU

A teraz po kolei na jednej warstwie podmieniamy funkcję aktywacji z sigmoidalnej na ReLU.

```python
def packmodel3(M, rel): # rel - która warstwa używa ReLU?
    M.add(Dense(15, input_dim=2, activation='sigmoid'))
    for i in range(14, 9, -1):
        if i == rel:
            M.add(Dense(i, activation='relu'))
        else:
            M.add(Dense(i, activation='sigmoid'))
    M.add(Dense(5, activation='softmax'))
    M.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
M_relu_layer = [Sequential() for x in range(5)]
for x in range(5):
    packmodel3(M_relu_layer[x], x+10)
    uczenie(M_relu_layer[x])
```

Opłaciło się wykorzystanie ReLU na warstwie o 12 neuronach, natomiast sieć z ReLU na 14 neuronach nauczyła się dużo gorzej od reszty.

### Zmiana liczby epok (jako parametru uczenia)

W kolejnym etapie wychodzimy ze zmian architektury i zmieniamy proces uczenia. Skupimy się na liczbie epok, ustawiając ją na wartości po kolei: 75, 150, 300, 600, 1200.

```python
def packmodel4(M): # tutaj bez zmian w architekturze
    M.add(Dense(15, input_dim=2, activation='sigmoid'))
    for i in range(14, 9, -1):
        M.add(Dense(i, activation='sigmoid'))
    M.add(Dense(5, activation='softmax'))
    M.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    
M_varep_layer = [Sequential() for x in range(5)]
for x in range(5):
    packmodel4(M_varep_layer[x])
    uczenie(M_varep_layer[x], ep=75*(2**x))
```

Bez zaskoczenia; im więcej epok, tym lepsze wyniki (ale i więcej czasu uczenia).

### Zmiana liczby podzbiorów

A teraz ustawiamy liczbę batchów na: 3, 6, 12, 24, 48.

```python
M_varbs_layer = [Sequential() for x in range(5)]
for x in range(5):
    packmodel4(M_varbs_layer[x])
    uczenie(M_varbs_layer[x], bs=3*(2**x))
```

Trudno tutaj wskazać jasnego lidera, ale skupię się na wynikach na części testowej i wyróżnię sieć uczoną z podziałem na 12 podzbiorów na epokę.

## Obserwacje

## Bonus: sieć doskonała

To znaczy, niekoniecznie doskonała, gdyż da się jeszcze lepsze, ale postanowiłem się pokusić o skonstruowanie sieci łączącej modyfikacje dające najlepsze wyniki. W pewnym stopniu dążymy do perfekcji.

```python
perfect = Sequential()
perfect.add(Dense(15, input_dim=2, activation='sigmoid'))
perfect.add(Dense(13, activation='sigmoid'))
perfect.add(Dense(12, activation='relu'))
perfect.add(Dense(11, activation='sigmoid'))
perfect.add(Dense(11, activation='sigmoid'))
perfect.add(Dense(10, activation='sigmoid'))
perfect.add(Dense(5, activation='softmax'))
perfect.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

uczenie(perfect, ep=1000, bs=12)
```

Oryginalnie było **95,28**% dla części treningowej, **93,6**% dla testowej, teraz mamy odpowiednio **97,36**% i **94,2**%. Udało się!
