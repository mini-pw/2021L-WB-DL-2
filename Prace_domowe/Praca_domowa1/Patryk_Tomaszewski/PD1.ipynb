```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dat_m = pd.read_csv("student-mat.csv",sep=";")
dat_p = pd.read_csv("student-por.csv",sep=";")
```

## Czy dostęp do internetu wpływa na ocenę końcową?

```python
sns.displot(dat_m, x="G3", col="internet", stat="density", common_norm=False).fig.suptitle("Matematyka")
sns.displot(dat_p, x="G3", col="internet", stat="density", common_norm=False).fig.suptitle("Portugalski")
plt.show()
```

Dla matematyki rozkłady są bardzo podobne, zatem hipotezę należy odrzucić.  
Natomiast dla portugalskiego widać duże różnice w rozkładzie, zatem korelacja może istnieć.

## Czy chęć podjęcia wyższej nauki wpływa na ocenę końcową?

```python
sns.displot(dat_m, x="G3", col="higher", stat="density", common_norm=False).fig.suptitle("Matematyka")
sns.displot(dat_p, x="G3", col="higher", stat="density", common_norm=False).fig.suptitle("Portugalski")
plt.show()
```

Tak jako można było się spodziewać, osoby wykazujące chęć podjęcia dalszej nauki uzyskują znacząco lepsze wyniki, niezależnie od przedmiotu.

## Czy istnieje zależność między ilością niezdanych klas a czasem spędzonym na naukę?

```python
sns.jointplot(data=dat_m, x="failures", y="studytime", kind="kde", fill=True).fig.suptitle("Matematyka")
sns.jointplot(data=dat_p, x="failures", y="studytime", kind="kde", fill=True).fig.suptitle("Portugalski")
plt.show()
```

Patrząc na wykresy nie ma podstaw aby odrzucić tą hipotezę.

## Czy istnieje zależność między relacjami w rodzinie a ilością spożywanego alkoholu?

```python
plotW = sns.jointplot(data=dat_m, x="famrel", y="Walc", kind="kde", fill=True)
plotW.fig.suptitle("W weekend")
plotW.fig.subplots_adjust(top=0.92)

plotD = sns.jointplot(data=dat_m, x="famrel", y="Dalc", kind="kde", fill=True)
plotD.fig.suptitle("W ciągu tygodnia")
plotD.fig.subplots_adjust(top=0.92)

plt.show()
```

Nie widać korelacji pomiędzy alkoholem spożywanym w weekend a relacją w rodzinie, jednak widać ją dla alkoholu spożywanego w ciągu tygodnia.
